```python
import gym
import gym_sokoban
import matplotlib.pyplot as plt

base_path = "/home/schk/1yr_report/latex/"
xs = []
es = ["Sokoban-v0", "Breakout-v4", "Qbert-v4", "SpaceInvaders-v4"]

for e in es:
    env = gym.make(e)
    env.reset()
    for _ in range(50):
        x, *_ = env.step(1)
    xs.append(x)

fig, axarr = plt.subplots(1, 4, figsize=(20,5))  # 1 row, 4 columns
for i in range(4):
    axarr[i].imshow(xs[i])
    axarr[i].axis('off')  # This will remove the ticks
plt.tight_layout()
plt.savefig(base_path+"Chapter2/Figs/"+"task.eps")
plt.show()
```
