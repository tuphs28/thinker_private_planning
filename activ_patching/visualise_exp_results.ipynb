```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
expname = "targetorder"
```

```python
expname = "shortcut"
fig, axs = plt.subplots(1,2, figsize=(10,4))
results_df = pd.read_csv(f"./results/{expname}/{expname}_allk__k1_layeralllayers_allsteps_cell_denoise.csv", index_col=0)
axs[0].bar(height=results_df.iloc[:,range(0,80)].mean(axis=1).sort_values()[-20:], x=range(20))
axs[0].set_xticks(range(20), labels=[f"l{a.split(':')[0][1]},c{a.split(':')[1].split('[')[1][:-2]}" for a in results_df.iloc[:,range(0,80)].mean(axis=1).sort_values()[-20:].index], rotation="vertical")
axs[0].set_title("Denoise")
axs[0].set_ylim(0,1)
results_df = pd.read_csv(f"./results/{expname}/{expname}_allk__k1_layeralllayers_allsteps_cell_noise.csv", index_col=0)
axs[1].bar(height=results_df.iloc[:,range(0,80)].mean(axis=1).sort_values()[-20:], x=range(20))
axs[1].set_xticks(range(20), labels=[f"l{a.split(':')[0][1]},c{a.split(':')[1].split('[')[1][:-2]}" for a in results_df.iloc[:,range(0,80)].mean(axis=1).sort_values()[-20:].index], rotation="vertical")
axs[1].set_title("Noise")
axs[1].set_ylim(0,1)
fig.suptitle(expname)
plt.show()
```

```python
expname = "targetorder"
fig, axs = plt.subplots(4,2, figsize=(10,15))
axs[0][0].set_title("Denoise")
axs[0][1].set_title("Noise")
for k in range(4):
    results_df = pd.read_csv(f"./results/{expname}/{expname}_allk__k1_layeralllayers_allsteps_cell_denoise.csv", index_col=0)
    axs[k][0].bar(height=results_df.iloc[:,range(k,80,4)].mean(axis=1).sort_values()[-20:], x=range(20))
    axs[k][0].set_xticks(range(20), labels=[f"l{a.split(':')[0][1]}c{a.split(':')[1].split('[')[1][:-2]}" for a in results_df.iloc[:,range(k,80,4)].mean(axis=1).sort_values()[-20:].index], rotation="vertical")
    axs[k][0].set_ylim(0,1)
    results_df = pd.read_csv(f"./results/{expname}/{expname}_allk__k1_layeralllayers_allsteps_cell_noise.csv", index_col=0)
    axs[k][1].bar(height=results_df.iloc[:,range(k,80,4)].mean(axis=1).sort_values()[-20:], x=range(20))
    axs[k][1].set_xticks(range(20), labels=[f"l{a.split(':')[0][1]}c{a.split(':')[1].split('[')[1][:-2]}" for a in results_df.iloc[:,range(k,80,4)].mean(axis=1).sort_values()[-20:].index], rotation="vertical")
    axs[k][1].set_ylim(0,1)
    fig.suptitle(expname)
plt.show()
```
